```python

# Step 1: Create a cross-tab of counts
cross_tab = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Step 2: Convert to percentage of total for each SeniorCitizen group
percent_df = cross_tab.div(cross_tab.sum(axis=1), axis=0) * 100

# Step 3: Plot
fig, ax = plt.subplots(figsize=(5,5))

# Plot each Churn category on top of each other (stacked)
bottom_vals = [0] * len(percent_df)
for churn_status in percent_df.columns:
    bars = ax.bar(percent_df.index,
                  percent_df[churn_status],
                  bottom=bottom_vals,
                  label=churn_status)
    
    # Add % labels
    for i, bar in enumerate(bars):
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width()/2,
                    bottom_vals[i] + height/2,
                    f'{height:.1f}%',
                    ha='center', va='center', color='white', fontsize=10, fontweight='bold')
    
    # Update bottom for next stacked section
    bottom_vals = [bottom_vals[i] + percent_df[churn_status].iloc[i] for i in range(len(bottom_vals))]

# Final touches
ax.set_title("Churn % by SeniorCitizen", fontweight='bold')
ax.set_xlabel("SeniorCitizen")
ax.set_ylabel("Percentage")
ax.legend(title="Churn")
plt.xticks([0, 1], ['Not Senior', 'Senior'])
plt.ylim(0, 110)
plt.tight_layout()
plt.show()
```

```python

# List of categorical columns
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Define subplot layout
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  # Ensures all plots fit

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 12))
axes = axes.flatten()  # Flatten 2D array of axes to iterate easily

# Loop through columns and create countplots
for i, col in enumerate(columns):
    sns.countplot(x=col, hue="Churn", data=df, ax=axes[i], palette='pastel', legend=False)
    axes[i].set_title(f'Countplot of {col}', fontsize=12, fontweight='bold')
    axes[i].tick_params(axis='x', rotation=45)

# Turn off any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Customers with no online security, backup, or tech support show higher churn rates, indicating these services help retain users.
# Fiber optic users churn more than DSL or those without internet.
# Streaming services (TV/movies) donâ€™t significantly reduce churn.
# Retention efforts should focus on boosting support and protection features.
```
